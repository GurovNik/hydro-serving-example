```python
import json
import urllib.request
import docker
import hydro_serving_grpc as hs
import grpc
import os
import base64

class ManagerWrapper:
    def __init__(self, hs_addr, docker_client):
        self.docker_client = docker_client
        self.hs_addr = hs_addr
    
    def post(self, url, data):
        """
        Sends POST request with `data` to the given `url`
        """
        json_data = json.dumps(data)
        req = urllib.request.Request(url, json_data.encode('utf-8'), {'Content-Type': 'application/json'})
        f = urllib.request.urlopen(req)
        response = json.loads(f.read().decode("utf-8"))
        f.close()
        return response
    
    def get(self, url):
        req = urllib.request.Request(url)
        f = urllib.request.urlopen(req)
        response = json.loads(f.read().decode("utf-8"))
        f.close()
        return response
    
    def add_runtime(self, name, version, rtypes, pull = False):
        image_name = "{}:{}".format(name, version)
        if pull:
            print("Pulling {} ...".format(image_name))
            self.docker_client.images.pull(image_name)
            
        print("Adding {} runtime ...".format(image_name))
        url = self.hs_addr + '/api/v1/runtime'
        data = {
              "name": name,
              "version": version,
              "modelTypes": rtypes,
              "tags": [],
              "configParams": {}
        }
        return self.post(url, data)
    
    def build_model(self, model_id):
        url = self.hs_addr + '/api/v1/model/build'
        data = {
          "modelId": model_id
        }
        return self.post(url, data)
    
    def add_application(self, application):
        url = self.hs_addr + '/api/v1/applications'
        return self.post(url, application)
    
    def serve_application(self, app_name, signature, data):
        url = self.hs_addr + '/api/v1/applications/serve/{0}/{1}'.format(app_name, signature)
        return self.post(url, data)
    
    def get_models(self):
        url = self.hs_addr + '/api/v1/model'
        return self.get(url)
    
    def create_app(self, app_def):
        pass
    
    def serve_app(self, app_name, serve_data):
        pass
```

```python
%matplotlib inline
from show_result import show_result

channel = grpc.insecure_channel('localhost:8080')
client = hs.PredictionServiceStub(channel=channel)

pics_dir = "/Users/bulat/Documents/Dev/SSD-Tensorflow/demo"
for file in os.listdir(pics_dir):
    full_file_path = os.path.join(pics_dir, file)
    
    if (not os.path.isfile(full_file_path)):
        print("Skipped: {}".format(full_file_path))
        continue
        
    print("Serving: {}".format(full_file_path))
    
    with open(full_file_path, "rb") as image_file:
        encoded_string = base64.b64encode(image_file.read())

    image_tensor = hs.TensorProto(
        dtype=hs.DT_STRING,
        string_val=[encoded_string]
    )

    request = hs.PredictRequest(
        model_spec=hs.ModelSpec(signature_name="detect", name=created_app["name"]),
        inputs={
            "image_b64": image_tensor,
        }
    )
    
    result = client.Predict(request)
    
    show_result(full_file_path, result)
```
