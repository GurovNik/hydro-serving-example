```python
samples_data = np.loadtxt("./HS_demo/claim_data/subset_large.csv", delimiter=",")
inliers_data = np.loadtxt("./HS_demo/claim_data/inliers.csv", delimiter=",")
outliers_data = np.loadtxt("./HS_demo/claim_data/outliers.csv", delimiter=",")

samples_data.shape, inliers_data.shape, outliers_data.shape
```

```python
def generator(data, size = 1):
    while True:
        mask = np.random.choice(range(len(data)), size)
        yield data[mask][0]
        
inliers = generator(inliers_data)
outliers = generator(outliers_data)

def infer_claims(gen):
    for _ in range(10000):
        data = next(gen)
        body = json.dumps({
            "client_profile": data.tolist()
        })

        headers = {'Content-Type': 'application/json'}
        r = requests.post(
            'http://demo.hydrosphere.io/api/v1/applications/serve/2/claim', 
            data = body,
            headers = headers
        )

        print(r.text[0:100], "...")
        time.sleep(0.6)
```

```python
infer_claims(inliers)
```

```python
infer_claims(outliers)
```
