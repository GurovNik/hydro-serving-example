```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def visualize_experiment(experiment_name: str):
    data = np.loadtxt(f"examples/{experiment_name}/data.txt", delimiter="\n")
    score = np.loadtxt(f"examples/{experiment_name}/score.txt", delimiter="\n")
    outlier = np.loadtxt(f"examples/{experiment_name}/flag.txt", delimiter="\n")
    fig, ax = plt.subplots(2)
    fig.set_size_inches((20, 5))

    outlier_color = np.where(outlier[:500], "red", "blue")
    ax[0].plot(data[:500])
    ax[0].scatter(y = data[:500], x = range(0, 500), c=outlier_color, s=7)
    ax[1].plot(score[:500])
    ax[1].scatter(y = score[:500], x = range(0, 500), c=outlier_color, s=10)
```

```python
visualize_experiment("sin_cos")
```

```python
visualize_experiment("sin_const")
```

```python
experiment_name = "three_clouds"
import matplotlib.cm as cm

data = np.loadtxt(f"examples/{experiment_name}/data.txt", delimiter="\n").reshape((1000,2))
score = np.loadtxt(f"examples/{experiment_name}/score.txt", delimiter="\n")
outlier = np.loadtxt(f"examples/{experiment_name}/flag.txt", delimiter="\n")
fig, ax = plt.subplots(2)
fig.set_size_inches((20, 10))

outlier_color = np.where(outlier[:], "red", "blue")
normalized_score = (score/ np.max(score))*100
ax[0].scatter(y = data[:, 0], x = data[:, 1], c=normalized_score, s=15, cmap='coolwarm')
ax[1].plot(score)
ax[1].scatter(y = score, x = range(0, 1000), c=outlier_color, s=10)
```
